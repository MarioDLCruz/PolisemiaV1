```python


import nltk
nltk.download('stopwords')
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords



def calcular_tf_idf_por_estado(rutas_archivos):
    tweets = []
    estados = []

    for ruta in rutas_archivos:
        try:
            with open(ruta, 'r', encoding='utf-8') as file:
                lines = file.readlines()
        except FileNotFoundError:
            print(f"Error: No se encontró el archivo en {ruta}")
            continue
        except Exception as e:
            print(f"Error al leer el archivo {ruta}: {e}")
            continue

        for linea in lines:
            if not linea.strip():
                continue

            partes = linea.split('_GEO', 1)
            if len(partes) < 2:
                continue

            estado = partes[0].strip()
            tweet = partes[1].strip().lower()

            tweets.append(tweet)
            estados.append(estado)

    tweets_df = pd.DataFrame({
        'Estado': estados,
        'Tweet': tweets
    })

    final_df = pd.DataFrame()
    stop_words_spanish = stopwords.words('spanish')
    for estado in tweets_df['Estado'].unique():
        tweets_estado = tweets_df[tweets_df['Estado'] == estado]['Tweet'].tolist()

        vectorizer = TfidfVectorizer(use_idf=True, min_df=5, stop_words=stop_words_spanish)
        tfidf_matrix = vectorizer.fit_transform(tweets_estado)

        palabras = vectorizer.get_feature_names_out()
        idfs = vectorizer.idf_
        tfs_totales = tfidf_matrix.sum(axis=0).A1  # Frecuencia total por término

        estado_df = pd.DataFrame({
            'Palabra': palabras,
            'Estado': estado,
            'IDF': idfs,
            'TF': tfs_totales
        })

        final_df = pd.concat([final_df, estado_df], ignore_index=True)

    return final_df,tweets_df,stop_words_spanish




rutas_archivos = [
    '/content/drive/MyDrive/state_twitter_2529_2022--03--14.txt',
    '/content/drive/MyDrive/state_twitter_2530_2022--03--18.txt',
    '/content/drive/MyDrive/state_twitter_2531_2022--03--20.txt',
    '/content/drive/MyDrive/state_twitter_2532_2022--03--22.txt',
    '/content/drive/MyDrive/state_twitter_2533_2022--03--24.txt',
    '/content/drive/MyDrive/state_twitter_2534_2022--04--03.txt',
    '/content/drive/MyDrive/state_twitter_2535_2022--04--12.txt',
    '/content/drive/MyDrive/state_twitter_2536_2022--04--17.txt',
    '/content/drive/MyDrive/state_twitter_2537_2022--04--22.txt',
    '/content/drive/MyDrive/state_twitter_2538_2022--04--23.txt',
    '/content/drive/MyDrive/state_twitter_2539_2022--04--27.txt',
    '/content/drive/MyDrive/state_twitter_2540_2022--04--28.txt',
    '/content/drive/MyDrive/state_twitter_2541_2022--04--29.txt',
    '/content/drive/MyDrive/state_twitter_2542_2022--04--30.txt',
    '/content/drive/MyDrive/state_twitter_2501_2021--10--11.txt',
    '/content/drive/MyDrive/state_twitter_2502_2021--10--12.txt',
    '/content/drive/MyDrive/state_twitter_2503_2021--10--13.txt',
    '/content/drive/MyDrive/state_twitter_2504_2021--10--16.txt',
    '/content/drive/MyDrive/state_twitter_2505_2021--10--17.txt',
    '/content/drive/MyDrive/state_twitter_2506_2021--10--19.txt',
    '/content/drive/MyDrive/state_twitter_2507_2021--10--31.txt',
    '/content/drive/MyDrive/state_twitter_2508_2021--11--10.txt',
    '/content/drive/MyDrive/state_twitter_2509_2021--11--12.txt',
    '/content/drive/MyDrive/state_twitter_2510_2021--11--22.txt',
    '/content/drive/MyDrive/state_twitter_2511_2021--11--24.txt',
    '/content/drive/MyDrive/state_twitter_2512_2021--11--27.txt',
    '/content/drive/MyDrive/state_twitter_2513_2021--12--03.txt',
    '/content/drive/MyDrive/state_twitter_2514_2021--12--05.txt',
    '/content/drive/MyDrive/state_twitter_2515_2021--12--20.txt',
    '/content/drive/MyDrive/state_twitter_2516_2021--12--21.txt',
    '/content/drive/MyDrive/state_twitter_2517_2021--12--26.txt',
    '/content/drive/MyDrive/state_twitter_2518_2021--12--27.txt',
    '/content/drive/MyDrive/state_twitter_2519_2021--12--30.txt',
    '/content/drive/MyDrive/state_twitter_2521_2022--01--03.txt',
    '/content/drive/MyDrive/state_twitter_2522_2022--01--15.txt',
    '/content/drive/MyDrive/state_twitter_2523_2022--01--26.txt',
    '/content/drive/MyDrive/state_twitter_2524_2022--02--03.txt',
    '/content/drive/MyDrive/state_twitter_2525_2022--02--12.txt',
    '/content/drive/MyDrive/state_twitter_2526_2022--03--02.txt',
    '/content/drive/MyDrive/state_twitter_2527_2022--03--12.txt',
    '/content/drive/MyDrive/state_twitter_2528_2022--03--13.txt',
    '/content/drive/MyDrive/state_twitter_2520_2021--12--31.txt',
]


resultado_df,tweets_df,stop_words_spanish = calcular_tf_idf_por_estado(rutas_archivos)


# Exportar a CSV
resultado_df.to_csv('tf_idf_por_estado.csv', index=False, encoding='utf-8')
```

Interseccion de todas las regiones

```python
def interseccion_tf_idf_por_estado(df_original):
    estados = df_original['Estado'].unique()

    # Crear sets de palabras por estado
    sets_palabras = {
        estado: set(df_original[df_original['Estado'] == estado]['Palabra'])
        for estado in estados
    }

    # Calcular intersección de todas las palabras entre estados
    palabras_comunes = set.intersection(*sets_palabras.values())


    df_final = pd.DataFrame({'Palabra': list(palabras_comunes)})

    # Añadir columna IDF_<estado> por cada estado
    for estado in estados:
        df_estado = df_original[df_original['Estado'] == estado]
        df_estado = df_estado[df_estado['Palabra'].isin(palabras_comunes)]
        df_estado = df_estado[['Palabra','IDF']].copy()
        df_estado.rename(columns={
            'IDF': f'IDF_{estado}'
        }, inplace=True)

        df_final = df_final.merge(df_estado, on='Palabra', how='left')

    return df_final
df_interseccion = interseccion_tf_idf_por_estado(resultado_df)
#df_interseccion.to_csv("interseccion_palabras_tf_idf.csv", index=False, encoding='utf-8')
```

Comparación entre zona norte y zona sur

```python
def comparar_idf_regiones(df, region_norte, region_sur):
    # Filtrar por cada región
    df_norte = df[df['Estado'].isin(region_norte)]
    df_sur = df[df['Estado'].isin(region_sur)]

    # Calcular IDF promedio por palabra en cada región
    idf_norte = df_norte.groupby('Palabra')['IDF'].mean().reset_index(name='IDF_Norte')
    idf_sur = df_sur.groupby('Palabra')['IDF'].mean().reset_index(name='IDF_Sur')

    # Hacer la intersección de palabras y unir ambos DataFrames
    df_comparado = pd.merge(idf_norte, idf_sur, on='Palabra', how='inner')

    # Calcular la diferencia
    df_comparado['Diferencia_IDF'] = df_comparado['IDF_Norte'] - df_comparado['IDF_Sur']

    # Ordenar por diferencia
    df_comparado.sort_values(by='Diferencia_IDF', ascending=False, inplace=True)

    return df_comparado
```

```python

norte = ['BC', 'BCS', 'Sonora']
sur = ['Chiapas', 'Guerrero', 'Oaxaca']

df_comparacion = comparar_idf_regiones(resultado_df, norte, sur)

```
